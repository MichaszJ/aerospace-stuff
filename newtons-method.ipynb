# Newton's Method

Assumptions for applications:

- Symbolic function will not be given
- Set of x and y points will be given

Useful links:

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
nums = 200
x_min = 0
x_max = 4
xnp = np.linspace(x_min, x_max, nums)
x = xnp.tolist()

# using function y = (x - 2)^2, expected root at x = 2
def y_func(x):
    return (x - 2)**2

y = [y_func(xi) for xi in x]
```

```python
# linear interpolation function
def interpolate(x, x0, x1, y0, y1):
    return y0 + (x - x0)*((y1 - y0)/(x1 - x0))
```

```python
x_n = random.uniform(x_min, x_max)
accuracy = 0.001
iterations = 0

print('Starting x: ', x_n)

plt.figure(dpi=150)
plt.plot(x, y)

while True and iterations < 150:
    for i, xi in enumerate(x):
        if x_n < xi and x_n > x[i-1]:
            x1 = xi
            x0 = x[i - 1]
            y1 = y[i]
            y0 = y[i - 1]
            break
            
    y_n = interpolate(x_n, x0, x1, y0, y1)
    
    plt.scatter(x_n, y_n, marker='.')
    
    if y_n < 0 + accuracy and y_n > 0 - accuracy:
        x_final = x_n
        break
    
    dy_n = (y1 - y0)/(x1 - x0)
    
    x_n = x_n - y_n/dy_n
    
    iterations += 1
    
print('Iterations: ', iterations+1)
print('Final x: ', x_final)
print('Final y: ', y_n)
plt.show()
```
